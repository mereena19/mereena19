# view the distribution of various features and its central tendencies

```python
#create histogram of feature 'Fixed Acidity'

sns.histplot(data=df,x='fixed acidity',color='orange',edgecolor='linen',alpha=0.5,bins=5)
plt.title('Histogram of fixed acidity')
plt.xlabel('Fixed Acidity')
plt.ylabel('Count')
plt.show()
```

Observations:
    The graph is normally distributed.Maximum values lies between 6 to 8.

```python
#Calculate its mean,median

mean=round(df['fixed acidity'].mean(),2)
mean
```

```python
median=df['fixed acidity'].median()
median
```

```python
#Create histogram and calculate mean,median

sns.histplot(data=df,x='fixed acidity',color='orange',edgecolor='linen',alpha=0.5,bins=5)
plt.title('Histogram of fixed acidity')
plt.xlabel('Fixed Acidity')
plt.ylabel('Count')

plt.vlines(df['fixed acidity'].mean(),ymin=0,ymax=4000,colors='blue', label='Mean')
plt.vlines(df['fixed acidity'].median(),ymin=0,ymax=4000,colors='red', label='Median')
plt.legend()
plt.show()
```

observation:mean and median is very close to each other. So, we can say that we can choose either of the parameters, say mean is the central tendency

```python
#create histogram of feature 'Volatile Acidity'

sns.histplot(data=df,x='volatile acidity',color='yellow',edgecolor='white',alpha=0.5,bins=5)
plt.title('Histogram of Volatile acidity')
plt.xlabel('Volatile acidity')
plt.ylabel('count')
plt.show()
```

observation: we can see that the data is ditributed towards right.As we have observed the skewness, we can find distribution using distplot.

```python
#plot distplot of feature 'Volatile acidity'

sns.distplot(df['volatile acidity'],color='yellow')
plt.title('Distplot of volatile acidity')
plt.xlabel('Volatile acidity')
plt.ylabel('density')
plt.show()
```

observation: Here mean and median are almost equal.It is normally distributed and is symmetric.

```python
#Calculate the skewness of volatile acidity

df['volatile acidity'].skew()
```

observation: Here value is greater than 1,so it is positively skewed.

```python
#Create histogram for feature'citric acidity'

sns.histplot(data=df,x='citric acid',color='green',edgecolor='linen',alpha=0.5,bins=5)
plt.title('Histogram of citric acidity')
plt.xlabel('Citric acidity')
plt.ylabel('Count')
plt.show()
```

Observation:It is skewed towards right and it is postively skewed.

```python
sns.histplot(data=df,x='citric acid',color='green',edgecolor='linen',alpha=0.5,bins=5)
plt.title('Histogram of citric acidity')
plt.xlabel('Citric acidity')
plt.ylabel('Count')
plt.vlines(df['citric acid'].mean(),ymin=0,ymax=4000,colors='blue',label='mean')
plt.vlines(df['citric acid'].median(),ymin=0,ymax=4000,colors='red',label='median')
plt.legend()
plt.show()
```

Observation:Mean and Median are almost equal.There is only small difference. so we can choose mean as measure of central tendency

```python
#Create countplot for feature 'quality'

sns.countplot(df['quality'])
plt.title('countplot of quality')
plt.xlabel('quality')
plt.ylabel('count')
plt.show()
```

```python
df['quality'].value_counts()
```

```python
#Calculate mode of feature quality

df['quality'].mode().iloc[0]
```

```python
#create a new pandas series called 'rep_acid' and contains details about reprentative qualities of diffrent types of acid

rep_acid=pd.Series(index=['fixed acidity','volatile acidity','citric acid','quality'],
                  data=[df['fixed acidity'].mean(),df['volatile acidity'].mean(),df['citric acid'].mean(),df['quality'].mode().iloc[0]])
```

```python
rep_acid
```

Observations:The mean value of fixed acidity is 6.8
The mean value of volatile acidity is 0.278
The mean value of citric acid is 0.33
The mean value of quality is 6
